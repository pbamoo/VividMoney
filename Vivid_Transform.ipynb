```python
#Import libraries
from pyforest import *
lazy_imports()
import warnings
warnings.filterwarnings('ignore')
```

```python
#import the dataset into pandas: 
data = pd.ExcelFile('Assignment.xlsx', engine = 'openpyxl')
```

```python
#Customers Monthly dataset
df_cust = pd.read_excel(data, 'Customers Monthly')
df_cust = df_cust[df_cust['CUSTOMER_HK'].notna()]
df_cust
```

```python
# Handling Missing Values
num_missing = df_cust.isnull().sum()
missing_per = (num_missing/df_cust.shape[0])*100
missing_table = pd.concat([num_missing,missing_per], axis=1, ignore_index=True) 
missing_table.rename(columns={0:'Num_Missing',1:'%_Missing'}, inplace=True)
print('\033[1m' + '\033[94m' + 'Missing Data Info:'.center(80) + '\033[0m')
print(missing_table)
```

```python
# Only ca. 3% of PLAN_NM is missing so we will remove those
df_cust.dropna(subset=['PLAN_NM'], inplace=True)
df_cust
```

```python
#For Conversations dataset

df_con = pd.read_excel(data, 'Conversations')
df_con = df_con[df_con['CUSTOMER_HK'].notna()] #remove rows with missing values in CUSTOMER_HK
df_con
```

```python
# Data Transformation

## Handling Missing Values
num_missing = df_con.isnull().sum()
missing_per = (num_missing/df_con.shape[0])*100
missing_table = pd.concat([num_missing,missing_per], axis=1, ignore_index=True) 
missing_table.rename(columns={0:'Num_Missing',1:'%_Missing'}, inplace=True)
print('\033[1m' + '\033[94m' + 'Missing Data Info:'.center(80) + '\033[0m')
print(missing_table)
```

```python
#convert to datetime format 
df_con['CREATED_AT'] = pd.to_datetime(df_con['CREATED_AT'].astype(str), format='%Y-%m-%d %H:%M:%S')
df_con['YEAR'] = df_con['CREATED_AT'].apply(lambda x: x.year)
df_con['CREATED_DATE'] = df_con['CREATED_AT'].dt.date
df_con['CREATED_YEAR'] = df_con['CREATED_AT'].dt.year
df_con['CREATED_MONTH'] = df_con['CREATED_AT'].dt.month
df_con['CREATED_MONTH_NAME'] = df_con['CREATED_AT'].dt.month_name()
df_con['CREATED_YEAR_MONTH'] = df_con['CREATED_AT'].dt.strftime('%Y-%m')
df_con['CREATED_DAY'] = df_con['CREATED_AT'].dt.day
df_con['CREATED_DAY_NAME'] = df_con['CREATED_AT'].dt.day_name()
df_con['CREATED_HOUR'] = df_con['CREATED_AT'].dt.hour
#df_con.set_index('date', inplace=True)
```

```python
#drop rows with NaNs in CLOSED_AT & FRT_DU_IN_MIN columns
df_conv = df_con.dropna(subset=['CLOSED_AT','FRT_DU_IN_MIN'])
```

```python
##Chat Handle Time (CHT)

# Create new column with calculated Chat Handle Time in minutes (CHT)
df_conv['CHT_MIN'] = df_conv['CLOSED_AT'] - df_conv['CREATED_AT']
df_conv['CHT_MIN']=df_conv['CHT_MIN']/np.timedelta64(1,'m')
```

```python
df_conv.columns
```

```python
#rearrange columns
df_conv = df_conv[['CUSTOMER_HK', 'CONVERSATION_HK', 'INITIATOR_CD', 'LANGUAGE_CD', 'CREATED_AT', 'CREATED_DATE', 
                   'CREATED_YEAR', 'CREATED_YEAR_MONTH', 'CREATED_MONTH', 'CREATED_MONTH_NAME', 'CREATED_DAY', 
                   'CREATED_DAY_NAME', 'CREATED_HOUR', 'CLOSED_AT', 'FRT_DU_IN_MIN', 'CHT_MIN', 'STATUS_CD', 'RATING_NUM']]
```
